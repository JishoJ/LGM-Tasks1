# Importing Packages

```python
import cv2
from google.colab.patches import cv2_imshow
import matplotlib.pyplot as plt
```

## Loading the Image

```python
image = cv2.imread("pic.jpg")
image = cv2.resize(image,(600,400))
cv2_imshow(image)
```

##Reading Image in RGB 

```python
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
plt.imshow(image_rgb)
plt.axis(False)
plt.show()
```

##Converting RGB format to Grey Scale image

```python
image_grey = cv2.cvtColor(image_rgb, cv2.COLOR_BGR2GRAY)
cv2_imshow(image_grey)
```

##Inverting the Grey Scale

```python
image_inverted = cv2.bitwise_not(image_grey)
cv2_imshow(image_inverted)
```

##Image Smoothing

```python
image_smoothing = cv2.GaussianBlur(image_inverted,(21,21),sigmaX=0, sigmaY=0)
cv2_imshow(image_smoothing)
```

##Pencil Sketch

```python
def pencil_sketch(x,y):
  return cv2.divide(x,255 - y,scale=256)
```

```python
image_pencil = pencil_sketch(image_grey,image_smoothing)
cv2_imshow(image_pencil)
```

##Displaying the Images

```python
cv2_imshow(image)
cv2_imshow(image_pencil)
```
